**Answers :**

PCA and NMF are two methods used in data analysis to reduce dimensions and extract features.

PCA:
- Allows both positive and negative values in its components.
- Creates components known as eigenfaces, which capture broad variations in data.
- Better for capturing overall structure and meaningful negative contributions from features.

NMF:
- Approximates data as non-negative linear combinations of basis components.
- Creates components, called NMF-faces, representing localized features like eyes or noses.
- Better for interpretability and strictly additive combinations.

PCA is faster and has closed-form solutions, making it efficient for large datasets.
NMF is iterative and computationally expensive, requiring more time.

In classification tasks, PCA may be better for capturing significant sources of variance, while NMF may excel at recognizing localized features.

Using PCA before NMF:
- Can reduce dimensionality and denoise data.
- May enhance interpretability of NMF components but can introduce negative values.
- Needs empirical validation for effectiveness in tasks like classification or clustering.

Overall, the decision to use PCA before NMF depends on the analysis goals and dataset characteristics.

````python
def NNMFLecture(X,r=None,N_Iter=1000,tolerance=1e-3,plot_evolution=1):
    '''
    Inputs: 
    %           X: is a [dxN] matrix. Every column (x) is an observation and every
    %           row consists of features.
    %
    %           r: size of the matrices W and H
    %
    %           (Optional) N_Iter: maximum number of iterations
    %
    %           (Optional) tolerance: convergence criteria threshold
    %
    %           (Optional) plot_evolution: plot evolution convergence criteria
    %
    % Outputs:
    %           W: is a [d x r] matrix containing the basis images in its
    %           columns
    %           
    %           H: is a [r x N] matrix containing the loadings (h) in its columns
    %           of the linear combination: x=Wh 
    %
  '''
    if r is None:
        r=X.shape[0]
        
    # Test for positive values
    if np.min(X) < 0:
        raise NameError('Input matrix X has negative values !')      

    # Size
    d,N=X.shape
   
    # Initialization
    W=np.random.rand(d,r)
    H=np.random.rand(r,N)   
    
    # parameters for convergence
    k = 0
    delta = np.inf
    eps=np.finfo(float).eps
    evolutionDelta=[]
 
    while delta > tolerance and k < N_Iter:
        
        # multiplicative method      
        
        XH=np.dot(X,H.T)
        HH=np.dot(H,H.T)
        for i in range(20):
            W = np.multiply(W, np.divide(XH, np.dot(W,HH) + eps))
                       
        H = np.multiply(H, np.divide(np.dot(W.T,X), np.dot(W.T,np.dot(W,H)) + eps))

        # Convergence indices
        k = k + 1           
        diff=X-np.dot(W,H)     
        #delta = np.sqrt(np.sum(diff**2)) / np.sqrt(np.sum(X**2)) # |X-WH|_2 / |X|_2
        delta = LA.norm(diff,'fro') / LA.norm(X,'fro') # sqrt(trace(diff'*diff)) / sqrt(trace(X'*X))
        evolutionDelta.append(delta)
        
        if k==1 or k%100==0:
            print('Iteration NMF number ', k, ' out of ', N_Iter , ', delta = ', delta, ', error (norm delta): ', LA.norm(diff))
     
    if k==N_Iter:
        print('Maximum number of iterations reached ! delta = ', delta)
    else:
        print('Convergence achieved ( delta = ', delta, ') in ', k, ' iterations')
    
    if plot_evolution==1:
        plt.figure(figsize=(6, 6))
        plt.plot(range(k),evolutionDelta,'bx--', linewidth=4, markersize=12)  
        plt.title('Evolution of error - NMF')
        plt.show()
    
    return W,H
````

## I-3) ICA and NMF in the presence of noise

We will now compare ICA and NMF results in the presence of noise within the dataset. Please note that during the class on ICA, the dataset $X$ was assumed to be noiseless: we were looking for a decomposition such that $X = AS$, and not $X = AS + N$, where $N$ would be a noise term.

1. How do you think that ICA might behave in the presence of noise? You might in particular consider the case of a Gaussian noise N.
2. On the other hand, do you think that NMF behaves better? If so, for what kind of noises?

**Answers :**

1) When Gaussian noise $ N $ is added to the dataset, the model becomes $ X = AS + N $:

    1. ICA is more robust to noise when the signal-to-noise ratio (SNR) is high. However, Gaussian noise can complicate ICA because it relies on non-Gaussian features for separation. Gaussian noise lacks these features, making component separation challenging.

    2. ICA's performance can significantly degrade in the presence of Gaussian noise, which lacks the higher-order statistical structure exploited by ICA. Extracted components may not accurately represent underlying signals and may be harder to interpret.

    3. ICA typically involves a pre-whitening step to make observed signals uncorrelated with unit variance. This step can amplify noise, especially if it's uncorrelated, as most Gaussian noise is.

2) NMF's robustness to noise depends on its characteristics:

    1. It handles additive, non-negative noise well, making it suitable for datasets with such noise patterns, like "clutter" or slight variations.

    2. With strictly positive Gaussian noise and a high signal-to-noise ratio, NMF can perform reasonably. However, if Gaussian noise contains negative values, violating NMF's non-negativity constraint, it could pose challenges.

    3. Negative values in noise can be set to zero, but this introduces bias. Incorporating regularization, like sparsity constraints, can enhance NMF's noise robustness by focusing on essential features and ignoring noisy ones.

    Variations like Convex-NMF or Sparse NMF can also be beneficial depending on noise characteristics and the application.

````python
def KMeansLecture(X,K=2,tol=1e-4):
    # Randomly initialize K centers from the data points
    np.random.seed(0)
    indices = np.random.choice(X.shape[0], K, replace=False)
    muTab = X[indices]

    muTab_prev = np.copy(muTab)

    while True:
        # Assign each data point to the nearest cluster center
        distances = np.sqrt(((X - muTab[:, np.newaxis])**2).sum(axis=2))
        yKmeans = np.argmin(distances, axis=0)

        # Update cluster centers based on the mean of data points assigned to each cluster
        for k in range(K):
            if np.sum(yKmeans == k) > 0:
                muTab[k] = np.mean(X[yKmeans == k], axis=0)

        # Check for convergence
        if np.linalg.norm(muTab - muTab_prev) < tol:
            break

        muTab_prev = np.copy(muTab)

    return yKmeans, muTab
````

3) In K-means, centroids are computed as interpolations of data points, not necessarily as actual data points themselves. To enforce centroids as original observations, we can use the K-medoids algorithm:

    1. **Initialization**: Start with \( K \) initial medoids, chosen randomly or by a heuristic.
    2. **Assignment Step**: Assign each point to the closest medoid using a distance metric like Euclidean.
    3. **Update Step**: Find the point within each cluster that minimizes the sum of distances to other points, making it the new medoid.
    4. **Repeat**: Iterate between assignment and update steps until convergence.

    ```python
    import numpy as np

    def KMedoids(D, K, max_iters=100):
        # D: distance matrix
        # K: number of clusters
        # Initialize medoids randomly
        medoids = np.random.choice(D.shape[0], K, replace=False)
        prev_medoids = np.zeros(K)
        iter = 0
        while not np.array_equal(medoids, prev_medoids) and iter < max_iters:
            iter += 1
            # Assignment step
            clusters = assign_to_closest_medoids(D, medoids)
            # Update step
            prev_medoids = medoids.copy()
            medoids = compute_new_medoids(D, clusters, K)
        return medoids, clusters

    def assign_to_closest_medoids(D, medoids):
        # Assign points to closest medoids
        clusters = {}
        for i in range(len(D)):
            distances = D[i, medoids]
            min_dist = np.argmin(distances)
            if min_dist in clusters:
                clusters[min_dist].append(i)
            else:
                clusters[min_dist] = [i]
        return clusters

    def compute_new_medoids(D, clusters, K):
        # Compute new medoids
        new_medoids = np.zeros(K, dtype=int)
        for k in range(K):
            cluster = clusters[k]
            if len(cluster) == 0:  # Empty cluster
                continue
            intra_cluster_distances = D[np.ix_(cluster, cluster)]
            cluster_sum = np.sum(intra_cluster_distances, axis=1)
            new_medoids[k] = cluster[np.argmin(cluster_sum)]
        return new_medoids

    medoids, clusters = KMedoids(D, K=3)
    ```

    K-medoids is robust to noise and outliers due to actual data points being medoids, but it can be computationally expensive compared to K-means.
